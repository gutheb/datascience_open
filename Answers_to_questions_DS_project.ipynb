Questions Data science mid-term project
1) Scikit-learn is an open-source package for predictive data analysis. It is built on NumPy, SciPy and matplotlib. It is also commercially usable and very accessible, but not ideal for very large data sets.
PyTorch is also an open-source python-based ML package. It has good GPU support but can require additional packages.
TensorFlow is another open-source ML package. It is good for very large data sets and deep learning, but not as efficient for smaller projects.
2) chEMBL is an open-source data base of bioactive molecules. chEMBL is accessible from the Web, or as a python package which is useful for ML. It is also possible to download the data base.
3) ML is a part of artificial intelligence that uses algorithms to learn from given data and then can make prediction on new and unseen data. This differs from traditional programming, because in the latter, every execution follows from a command (visible in the code) and this is not the case in ML.
4) The main techniques in ML are supervised learning which uses input data that already contains the correct output attached as information, unsupervised learning which analyses unlabelled datasets to discover patterns and structure and reinforcement learning in which the machine receives feedback for its actions.
5) The different types of ML algorithms include classification and regression algorithms, clustering, dimensionality reduction and neural networks.
6) Common application of ML includes analysis of large datasets to understand structure and properties of molecules (like what we are trying to achieve), personalize online experiences on websites, predict stock market and even analyse medical images to detect diseases.
7) You can give it new values to which you know the wanted outcomes and compare the results, essentially the model passes a test.
8) It is important to have clean data. This means deleting duplicates, empty cells or wrong data. Also, it is important to structure the data such that it remains clear what has already been used by the model for training and what has not.
9) The model can become too specific and performs well in the data set of the training but badly outside. This issue can be solved with cross validation. The opposite can also happen, where the model not advanced enough to detect any kind of structure or pattern. Then the model needs more training and become more complex.
10) Some ways to practice ML is by using the libraries mentioned above, as they are accessible and by doing some practice in the course on Jupyter notebooks.


Questions on the data Set:
1)  The Size of the training set is 179827 rows and 5 columns. It contains the chembl ID which is a unique ID for every molecule in chembl, standard value which is the measured activity value of the molecule, unit indicates that the measurements are in nM, it also contains the biological target of each molecule and finally the SMILES which represents the chemical structure of the molecule.

2) The dataset was filtered to keep only the SMILES and activity values, missing values have been removed, non-positive activity values were excluded, and the activity values were transformed into pIC50. The cleaned dataset containing smiles and pIC50 was then saved as a CSV file.

3) A test set is a part of the data used to evaluate the performance of the trained model using unseen data. It is not used during training. We split the dataset is split into 70% training data and 30% test data. Other types of datasets include the training set, used to train the model, and the validation set, used to fine tune.

4) The first cell (90) imports the libraries used in the project. The cells 90-92 set the file path, load the dataset from a CSV file into a DataFrame, reset the index, and shows the shape and summary information to check its size and missing values. The next cell (93) cleans the dataset by keeping only the SMILES and activity columns, removing missing or non-positive values, converting the activity values to pIC50, and saving the dataset as a new CSV file. Also, it shows the first few rows of the set. Cells 95-97 sample 10,000 molecules from the cleaned dataset, define a function to convert SMILES strings into molecular fingerprints and displays the first few rows of the dataset. Cells 98,99 saves the full dataset with fingerprints as a CSV file, splits the data into training and test sets, and saves these sets as separate CSV files for model training and evaluation. Cells 100-101 define a connected neural network model with two hidden layers using ReLU activation and a linear output layer for predicting pIC50. Cell 102 converts the training and test fingerprint data into NumPy arrays, then trains the neural network with different batch sizes. It plots the training and validation loss over epochs to see model performance and compare how batch size affects learning. Cells 103-104 save the best model weights using a checkpoint during training and fits the model on the full training set with a batch size of 16. It also evaluates its performance on the test set by showing the loss, mean squared error, and mean absolute error. Cells 105-106 use the trained model to predict pIC50 values on the test set and create a scatter plot comparing predicted versus true pIC50 values to see if the model is good. Cells 107-108 load external data for to test the model on other kind of data. Cells 109-110 rebuilds the neural network architecture, loads the previously saved best weights and is used to predict pIC50 values for new dataset. Cells 111-112 save the predicted pIC50 values for the external dataset to a CSV file and selects the top three molecules with the highest predicted pIC50 values, identifying the best potential inhibitors. Cell 113 draws the molecules.
