 #### 1) Load Anaconda on UBELIX
 > module load Anaconda3
 > eval "$(conda shell.bash hook)"
 #This initializes conda so you can create and activate environments in the current shell

#### 2) Create a new environment (only i not yet created)
 > conda create -n <venv> python=3.10
 #call venv however you like i.e: kinaseenv

#### 3) Activate it
 > conda activate kinaseenv
#now you see (venv) in front of > 

#### 4) Install your packages
 (venv) >
 > conda install -c conda-forge pandas numpy matplotlib seaborn scikit-learn rdkit
 > pip install tensorflow

rdkit is commonly installed from conda-forge, and conda is the recommended route for scientific packages; pip can then be used inside the environment for packages not available via conda. If tensorflow gives you trouble with pip, we can adjust the exact version depending on your UBELIX Python version and GPU setup.

#### 5) Test the environment
 > python -c "import pandas, numpy, rdkit, sklearn, matplotlib, seaborn, tensorflow; print('all good')"

If that works, the environment is ready.
Keep it for later runs
